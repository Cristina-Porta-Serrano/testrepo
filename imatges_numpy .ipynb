##*Exercice 3*##

*Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.*

```python
matrix = np.random.randint(1,100,size=(5,5))
print(matrix)
```

```python
max_value = max([max(i) for i in matrix])
print(max_value)
```

```python
columnMaxValue = np.amax(matrix, axis=0)
print('Maximum value of the matrix along columns is:')
print(columnMaxValue)
```

```python
rowMaxValue = np.amax(matrix, axis=1)
print('Maximum value of the matrix along rows is:')
print(rowMaxValue)
```
